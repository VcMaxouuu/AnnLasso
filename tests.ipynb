```python
import numpy as np
from harderLASSO.models import harderLASSORegressor, harderLASSOClassifier
from compared_models.classification import LNClassifier
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

# Linear regression

```python
X = np.random.normal(size=(100, 300))
s = 20
features = np.arange(s)
beta = 3*np.ones(s)
y = X[:, features]@beta + np.random.normal(size=(X.shape[0],))

model = harderLASSORegressor()
model.fit(X, y, 1)
print(np.sqrt(mean_squared_error(y, model.predict(X))))
print(model.imp_feat)
```

# Non linear regression

```python
X = np.random.normal(size=(5000, 50))
y = 10*np.abs(X[:, 0]-X[:, 1]) + 10*np.abs(X[:, 2]-X[:,3]) + 10*np.abs(X[:, 4]-X[:, 5]) + 10*np.abs(X[:, 6]-X[:, 7]) + np.random.normal(size=(X.shape[0], ))

model = harderLASSORegressor(hidden_dims=(20, ), nu=0.1)
model.fit(X,y,1)
print(np.sqrt(mean_squared_error(y, model.predict(X))))
print(model.imp_feat)
```

# Classification

```python
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score

X,y = load_wine(return_X_y=True)
model = harderLASSOClassifier(hidden_dims=(20, ))

model.fit(X,y)
print(accuracy_score(model.predict(X), y))
print(model.imp_feat)
```
